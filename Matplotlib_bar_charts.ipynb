Matplotlib - Bar charts

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing Microsoft Data

csv_path = ('/Users/antonioelias/Desktop/MOOCs/Python Coding Bootcamp/Data Science Libs/microsoft.csv')
microsoft_data = pd.read_csv(csv_path)
microsoft_data.head()
```

```python
# Summary Statistics

microsoft_data.describe()
```

```python
# Creating Series

date = microsoft_data.loc[0:15, 'Date']
open = microsoft_data.loc[0:15, 'Open']
close = microsoft_data.loc[0:15, 'Close']
high = microsoft_data.loc[0:15, 'High']
```

```python
# Stacked bars 

plt.figure(figsize=(10, 6))
plt.bar(date, open, label='Open')
plt.bar(date, close, label='Close')
plt.bar(date, high, label='High')

plt.title('Microsoft Data - Stacked Bars')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tick_params(axis='x', rotation=70)

plt.legend()
plt.show()
```

```python
# Side-by-side bar chart

# Set bar width
w = 0.20

# Indexing Date (Used to manipulate spacing/location)
date_index = np.arange(len(date))

open_loc = (date_index - w)
close_loc = (date_index + w)
high_loc = (date_index)

plt.figure(figsize=(10, 6))

plt.bar(open_loc, open, width=w, label='Open')
plt.bar(close_loc, close, width=w, label='Close')
plt.bar(high_loc, high, width=w, label='High')

plt.title('Microsoft Data - Side-by-side bar chart')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tick_params(axis='x', rotation=70)

# Returning dates 
plt.xticks(ticks=date_index, labels=date)
plt.legend()

plt.show()
```
