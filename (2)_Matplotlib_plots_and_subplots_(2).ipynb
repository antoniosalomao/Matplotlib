```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Microsoft Data

csv_path = ('/Users/antonioelias/Desktop/MOOCs/Python Coding Bootcamp/Data Science Libs/microsoft.csv')
microsoft_data = pd.read_csv(csv_path)
microsoft_data.head(10)
```

```python
# Summary Stats
 
microsoft_data.describe()
```

```python
# Creating individual series and plotting them using Matplotlib

open_data = microsoft_data.loc[1:10, 'Open']
high_data = microsoft_data.loc[1:10, 'High']
low_data = microsoft_data.loc[1:10, 'Low']
date_data = microsoft_data.loc[1:10, 'Date']

plt.figure(figsize=(10, 6))
plt.title('Microsft Data')
plt.ylabel('Price (USD)')
plt.xlabel('Date')

plt.plot(date_data, open_data, 'r--', label='Open')
plt.plot(date_data, high_data,'g-', label='High')
plt.plot(date_data, low_data, 'c', label='Low')
plt.legend()
plt.tick_params(axis='x', rotation=70)
plt.show()


```

```python
# Subplots

# Creating two subplots in the same figure 
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,6))

# Subplot 1
ax1.plot(date_data, open_data, 'r--', label='Open')
ax1.plot(date_data, high_data, 'g-', label='Low')

ax1.set_xlabel('Date')
ax1.set_ylabel('Price (USD)')
ax1.tick_params(axis='x', rotation=70)

# Subplot 2
ax2.plot(date_data, low_data, 'c', label='Low')

ax2.set_xlabel('Date')
ax2.set_ylabel('Price (USD)')
ax2.tick_params(axis='x', rotation=70)

ax1.legend()
ax2.legend()

plt.show()

```

```python
# Extra (Exercise)

array = np.arange(0, 10, 0.1)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,6))

ax1.plot(array, np.sin(array), 'b--', label='Sin')
ax2.plot(array, np.sin(array)*-1, 'k', label='(-)Sin')

ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-label')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-label')

ax1.set_title('Sin')
ax2.set_title('(-)Sin')

ax1.legend()
ax2.legend()

plt.show()
```
