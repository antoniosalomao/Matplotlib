Matplotlib - Pie Chart

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Creating data 
val_list = [10e6, 20e6, 30e6, 40e6]
eq_list = ['Microsoft', 'Apple', 'Amazon', 'Zoom']

# Transfering data into a dictionary
my_dict_2 = {}
for i, j in enumerate(val_list):
    key = eq_list[i]
    my_dict_2[key] = j

# "Exploding" values (second element)
expl = [0, 0.1, 0, 0]

plt.figure(figsize=(6, 6))
plt.pie(my_dict_2.values(), labels=my_dict_2.keys(), shadow=True, autopct='%1.0f%%', explode=expl, wedgeprops={'edgecolor':'black'})

plt.title('Volume')

plt.show()
```
