```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import numpy as np 

array = np.arange(0, 10, 0.1)
plt.plot(array)

# Note that the array values are passed on to the y-axis
```

```python
sin = np.sin(array)
plt.plot(sin*-1)
plt.plot(sin)
```

```python
plt.figure(figsize=(10,6))
plt.plot(sin)
plt.plot(sin*-1)
plt.savefig('output.png')
```

```python
plt.figure(figsize=(10,6))
plt.title('Sin Function applied to Array (0, 10, 0.1)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(sin)
plt.plot(sin*-1)
```

```python
array_1 = sorted(np.random.randint(0, 50, 10))
array_2 = np.random.randint(0, 50, len(array_1))
plt.figure(figsize=(10, 6))
plt.title('Random Plot (Noise?')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(array_1, array_2)
plt.plot([1, 2, 3, 4], [0, 15, 7, 10])
```
