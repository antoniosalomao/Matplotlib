Matplotlib - Scatter Plots

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline

# Importing Microsoft Data 

csv_path = ('/Users/antonioelias/Desktop/MOOCs/Python Coding Bootcamp/Data Science Libs/microsoft.csv')
microsoft_data = pd.read_csv(csv_path)
microsoft_data.head()
```

```python
# Summary Statistics 

microsoft_data.describe()
```

```python
# Creating Series

date = microsoft_data.loc[0:15, 'Date']
open = microsoft_data.loc[0:15, 'Open']
volume = microsoft_data.loc[0:15, 'Volume']
```

```python
# Scatter PLot - V1 (Colorbar)

plt.figure(figsize=(10, 6))
plt.scatter(date, open, c=volume, cmap='summer', label='Volume')

plt.title('Microsoft Data - Scatter Plot - Colorbar')
plt.ylabel('Price (USD)')
plt.xlabel('Date')
plt.tick_params(axis='x', rotation=70)

cbar = plt.colorbar()
cbar.set_label('Volume')

plt.show()
```

```python
# Scatter Plot - V2 (Scalar)

# Normalizing volume
volume = volume/volume[0]

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(date, open, s=volume*30, label='Volume')

plt.title('Microsoft Data - Scatter Plot - Scalar')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.tick_params(axis='x', rotation=70)
plt.legend()
plt.show()
```
